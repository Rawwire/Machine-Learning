```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file=pd.read_csv("D:\Acmegrade\Files\Mall_Customers.csv")
file
```

```python
new_data=file.iloc[:,3:5]
```

```python
new_data
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5, init="k-means++")
kmeans.fit(new_data)
```

```python
kmeans.labels_
```

```python
pd_data=pd.DataFrame(kmeans.labels_,columns=["cluster No"])
result=pd.concat([file,pd_data],axis=1)
result
```

```python
result.to_excel("D:\Acmegrade\Files\mall.xlsx")
```

```python
new_data["cluster"]=kmeans.labels_
sns.set_style("whitegrid")
sns.lmplot(x="Annual Income (k$)", y="Spending Score (1-100)", data=new_data,hue='cluster',height=10,fit_reg=False)
```
