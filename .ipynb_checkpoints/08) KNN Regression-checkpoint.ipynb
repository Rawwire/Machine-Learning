```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file= pd.read_csv("D:\\Acmegrade\\Files\\50_Startups.csv")
```

```python
x= file.iloc[:,:-1].values
y=file.iloc[:,-1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder as ohe
from sklearn.preprocessing import LabelEncoder as le
```

```python
le=le()
x[:,3]=le.fit_transform(x[:,3])
x
```

```python
ct=ColumnTransformer(transformers=[("onehot",ohe(),[3])],remainder="passthrough")
x=ct.fit_transform(x)
x
```

```python
import statsmodels.api as sm
x=x.astype("float64")
x_opt=x[:,[0,1,2,3]]
a=sm.OLS(endog=y, exog=x_opt)
a=a.fit()
print(a.summary())
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test= tts(x_opt,y,test_size=0.2)
```
