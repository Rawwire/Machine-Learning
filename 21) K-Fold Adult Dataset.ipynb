```python
file["marital-status"]=file["marital-status"].replace(["Married-AF-spouse","Married-civ-spouse","Married-spouse-absent"],0)
file["marital-status"]=file["marital-status"].replace(["Divorced","Never-married","Separated","Widowed"],0)
```

```python
file.drop(labels=["workclass","education","occupation","relationship","race","native-country"],axis=1, inplace=True)
```

```python
file
```

```python
x=file.iloc[:,0:-1].values
y=file.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
models=[]
```

```python
models=[]
models.append(('Lr',LogisticRegression()))
models.append(("Knn",KNeighborsClassifier()))
models
```

```python
result=dict()
for name,model in models:
    kfold=KFold(n_splits=10)
    cv_results=cross_val_score(model,x_train,y_train,cv=kfold, scoring="accuracy")
    result[name]= (cv_results, cv_results.mean(), cv_results.std())
```

```python
result
```

```python
for key,values in result.items():
    print("Model_Name:",key, "\n Indivual_Accuracy:",values[0], "\n Mean:",values[1], "\n Standard_Deviation:", values[2])
```
