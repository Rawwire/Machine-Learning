```python
import numpy as np
import pandas as pd
```

```python
file=pd.read_csv("D:\\Acmegrade\\Files\\Boost_mushrooms.csv")
file
```

```python
file.info()
```

```python
from sklearn.preprocessing import LabelEncoder
for label in file.columns:
    file[label]=LabelEncoder().fit_transform( file[label])
```

```python
file
```

```python
y= file.iloc[:,0].values
x=file.iloc[:,2:].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
adaboost=AdaBoostClassifier(base_estimator=lr, n_estimators=400, learning_rate=1)
ad=adaboost.fit(x_train,y_train)
print(ad)
```

```python
df_pred=ad.predict(x_test)
```

```python
df_pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,df_pred)
acc
```

```python
from sklearn.tree import DecisionTreeClassifier as dtc
dt=dtc()
ad1=AdaBoostClassifier(base_estimator=dt, n_estimators=400, learning_rate=1)
ad1
```

```python
d_pred2=ad1.fit(x_train,y_train)
df_pred2=d_pred2.predict(x_test)
df_pred2
```

```python
acc=accuracy_score(y_test,df_pred2)
acc
```
