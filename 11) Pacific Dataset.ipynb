```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import random
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
pd.set_option("display.width",5000)
file=pd.read_csv("D:\Acmegrade\Files\pacific.csv")
file
        
```

```python
print(file.groupby("Status").agg({"Status":np.size}))
```

```python
file.Status=pd.Categorical(file.Status)
file["Status"]=file.Status.cat.codes
```

```python
file
```

```python
file_temp=file[:]
file_temp
```

```python
file_temp.drop(["ID"],axis=1,inplace=True)
file_temp.drop(["Name"],axis=1,inplace=True)
file_temp.drop(["Date"],axis=1,inplace=True)
file_temp.drop(["Time"],axis=1,inplace=True)
file_temp.drop(["Event"],axis=1,inplace=True)
file_temp.drop(["Status"],axis=1,inplace=True)
file_temp.drop(["Latitude"],axis=1,inplace=True)
file_temp.drop(["Longitude"],axis=1,inplace=True)
```

```python
file_temp
```

```python
x=file_temp.values
x
```

```python
y=file["Status"].values
y
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test= tts(x,y, test_size=0.4)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score as asc
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr_pred=lr.predict(x)
acc_lr=asc(y,lr_pred)
print(acc_lr)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier(n_neighbors=100, metric="minkowski", p=2)
knc.fit(x_train,y_train)
knc_pred=knc.predict(x)
acc_knc=asc(y,knc_pred)
print(acc_knc)
```
