```python
x= file.iloc[:,:-1].values
y=file.iloc[:,-1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder as ohe
from sklearn.preprocessing import LabelEncoder as le
```

```python
le=le()
```

```python
x[:,3]=le.fit_transform(x[:,3])
```

```python
x
```

```python
ct=ColumnTransformer(transformers=[("onehot",ohe(),[3])],remainder="passthrough")
x=ct.fit_transform(x)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler as ss
```

```python

x[:,3:5]=ss.fit_transform(x[:,3:5])
```

```python
x
```

```python
print(x.dtype)
```

```python
x=x.astype("float64")
```

```python
print(x.dtype)
```

```python
import statsmodels.api as sm
```

```python
a=sm.OLS(endog=y, exog=x)
a=a.fit()
print(a.summary())
```

```python
x_opt=x[:,[0,1,2,3]]
a=sm.OLS(endog=y, exog=x_opt)
a=a.fit()
print(a.summary())
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test= tts(x_opt,y,test_size=0.2)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression 
lr= LinearRegression()
lr.fit(x_train,y_train)
```

```python
dif_pred=lr.predict(x_test)
```

```python
dif_pred
```

```python
from sklearn.metrics import r2_score
acc=r2_score(y_test,dif_pred)
acc
```

```python
final=lr.predict(x_opt)
```

```python
final
```

```python
acc=r2_score(y,final)
```
