```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file=pd.read_csv("D:\Acmegrade\Files\mtcars.csv")
file
```

```python
x=file.iloc[:,[1,3,5,6]].values
x
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.rcParams['figure.figsize']=(20,10)
dendrogram=sch.dendrogram(sch.linkage(x, method="ward"))
```

```python
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=2, affinity="euclidean", linkage="ward")
acy=ac.fit_predict(x)
acy
```

```python
print(pd.DataFrame(acy).value_counts())
```

```python
plt.scatter(x[acy==0,0], x[acy==0,1], s=100, color="red", label="cluster1")
plt.scatter(x[acy==1,0], x[acy==1,1], s=100, color="green", label="cluster2")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(x, acy)
silhouette_avg
```
